## **Placeholder -** Migrate tensorflow 1 code to tensorflow 2

Placeholder is a variable that we will assign data to at a later date.

```python
# Turn off the modern line-by-line execution (Eager Execution) of TensorFlow 2.x
tf.compat.v1.disable_eager_execution()
```

```python
# Create an empty placeholder (a promise to provide data later)
tf.compat.v1.placeholder(dtype = tf.float32, shape=(400,400))
```

```python
# Create a placeholder blueprint and store it in variable 'a'
a = tf.compat.v1.placeholder(dtype = tf.float32, shape = (400,400))
a
```

```python
# Create a placeholder blueprint and store it in variable 'b'
b = tf.compat.v1.placeholder(dtype=tf.float32, shape=(400,400))
b
```

```python
# Define the addition operation between 'a' and 'b'
c = tf.add(a,b)
c
```

```python
# Create a 400x400 matrix filled entirely with the number 1.0
ones_array = np.ones((400,400), np.float32)
ones_array
```

```python
# Start the TensorFlow execution engine (Session)
with tf.compat.v1.Session() as sess:
  # Run the operation 'c' (which was a + b) and feed the actual data into the placeholders
  d = sess.run(c, feed_dict = {a:ones_array, b:ones_array})
```

```python
d
```

```python
# Create or retrieve a variable named "x" in the TensorFlow graph
# We explicitly set its type to 32-bit integer and initialize it with the value [5]
x = tf.compat.v1.get_variable("x", dtype=tf.int32, initializer=tf.constant([5]))

# Create or retrieve another variable named "y"
# Initialized with the value [6]
y = tf.compat.v1.get_variable("y", dtype=tf.int32, initializer=tf.constant([6]))
```

```python
# Create two constants and assign them specific names in the computational graph
# 'name' parameter helps in tracking these nodes if we visualize the graph later
c = tf.constant([5], name = "constant")
square = tf.constant([2], name = "square")

# Define the mathematical equation: f = (x * y) + x^2 + y + c
# The operations are added to the TensorFlow blueprint, but not calculated yet
f = tf.multiply(x,y) + tf.pow(x, square) + y + c

f
```

### Output Observation :
- **No Calculation Yet:** Because Eager Execution is turned off, the engine has only connected all the mathematical steps into a chain (graph), but has not yet flowed the actual values ​​inside the variables x and y.

- **How to execute this? :** To get the actual number 66 from Is f, you will need to start a Session. But a new rule applies here: variables, like placeholders, are not automatically ready to run. Just before running the session you need to turn on a master switch in the graph called tf.compat.v1.global_variables_initializer(), only then will x and y release their initial values ​​(5 and 6) in memory.
