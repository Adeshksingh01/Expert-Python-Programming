**import ABC from abc**

**abc -** stand for Abstract Base Classes

```python
# abc stand for Abstract Base Classes
from abc import ABC, abstractmethod
```

**Questions 1:**

```python
# Create a abstract class for Animal.
class Animal(ABC):
  @abstractmethod
  def make_sound(self):
    pass

# subclass inherite from the abstract class
class Dog(Animal):
  def make_sound(self):
    print("Bark")

class Cat(Animal):
  def make_sound(self):
    print("Meow")

# Creating instances of subclasses
dog = Dog()
cat = Cat()
```

```python
dog.make_sound()
```

```python
cat.make_sound()
```

**Question 2:**

```python
# Create a abstract class for Car.
class Car(ABC):
  @abstractmethod
  def mileage(self):
    pass
  def color(self):
    print("White")

# subclass inherite from the abstract class
class Suzuki(Car):
  def mileage(self):
    print("The mileage is 25km/h")

class Honda(Car):
  def mileage(self):
    print("The mileage is 24km/h")

# Creating instances of subclasses
suzuki = Suzuki()
honda = Honda()
```

```python
suzuki.mileage()
```

```python
suzuki.color()
```

```python
honda.mileage()
```

```python
honda.color()
```

**Question 3:**

```python
# Create a abstract class for Vehicle.
class Vehicle(ABC):
  @abstractmethod
  def max_speed(self):
    pass

# subclass inherite from the abstract class
class Car(Vehicle):
  def max_speed(self):
    return "200 Km/H"

class Bike(Vehicle):
  def max_speed(self):
    return "100 Km/H"

# creating instances of subclass
car = Car()
bike = Bike()
```

```python
car.max_speed()
```

```python
bike.max_speed()
```

**Question 4:**

```python
# Create a abstract class for Bankapp.
class BankApp(ABC):
  def database(self):
    print("Connected to database")

  @abstractmethod
  def security(self):
    pass

# subclass inherite from the abstract class
class mobileApp(BankApp):
  def mobile_login(self):
    print("Login into mobile")

  def security(self):
    print("Mobile Security")
```

```python
mob = mobileApp()
mob.database()
mob.mobile_login()
mob.security()
```

**Question 5:**

```python
# Create a abstract class for Shape.
class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

# Subclass inherite from the abstract class
class Cricle(Shape):
  def __init__(self, radius):
    self.radius = radius

  def area(self):
    return 3.17 * self.radius * self.radius
```

```python
cricle = Cricle(5)
print(cricle.area())
```
