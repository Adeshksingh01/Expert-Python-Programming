```python
class Bird:
  def fly(self):
    print("Bird can fly")

class Penguin(Bird):
  def fly(self):
    print("Penguin can't fly")

b = Bird()
p = Penguin()

b.fly()
p.fly()
```

Question : Write a python program to demonstrate polymorphism where two different classes have the same method name, and a common functions calls those method.

```python
class Cat:
    def sound(self):
      return "Meow"

class Dog:
    def sound(self):
      return "Woof"

cat = Cat()
dog = Dog()

# Using Polymorphism
animals = [cat,dog]
for animal in animals:
    print(animal.sound())
```

Question : Create a base class Vehicle with a method speed(). Then, create a derived class Car that overrides the speed() method.

```python
class Vehicle:
  def speed(self):
    print("The Vehicle move at standard speed.")

class Car(Vehicle):
  def speed(self):
    print("The car move at a speed of 120 km/h.")

vehicle = Vehicle()
car = Car()

vehicle.speed()
car.speed()
```
