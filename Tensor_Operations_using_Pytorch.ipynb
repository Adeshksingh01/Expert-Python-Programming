#### Creating Tensor..

```python
# Using empty
# Creates a 2*3 tensor (2 rows, 3 columns) but does not initialize the memory.
# It will contain whatever random "garbage" values were already sitting in your Computer's RAM.
a = torch.empty(2,3)
print(a)
```

```python
# Check type
# The type() function checks the python class/data structure of the variable "a".
type(a)
```

```python
# Using zeros
# Creates a 2*3 tensor (2 rows, 3 columns) completely filled with zeros (0.0)
# By default, PyTorch creates these zeros as float (decimal) numbers, not integers.
torch.zeros(2,3)
```

```python
# Using ones
# Creates a 2*3 tensor completely filled with ones(1.0)
# By default, PyTorch creates these ones as float (decimal) numbers, not integers.
torch.ones(2,3)
```

```python
# Using rand
# Creates a 4*3 tensor (4 rows and 3 columns) filled with random decimal numbers.
# The numbers are chosen from a "uniform distribution" between 0.0 and 1.0.
# It return a float data type.
torch.rand(4,3)
```

```python
# Using manual_seed
# Lock the random number generator with a specific "key" or seed (e.g., 500)
torch.manual_seed(500)

# Generates a 2*3 tensor of random decimal numbers.
# Because of the seed, these specific numbers will NEVER changes upon rerunning.
torch.rand(2,3)
```

```python
# Using tensor
# Converts a standard Python list of lists into a PyTorch Tensor.
# Since we provided integers (1,2,3..), PyTorch automatically infers the type as int64.
t = torch.tensor([[1,2,3],[4,5,6]])
print(t)
print("Datatype:",t.dtype)
```

```python
# Using arange
# Create a sequence from start (0) to end (10, exclusive) with a step size (2).
a = torch.arange(0,10,2)
print("Using arange ->",a)
```

```python
# Using linspace
# Creates "7" equally spaced points between start (0) and end (10, inclusive).
l = torch.linspace(0,10,7)
print("Using linspace ->", l)
```

```python
# Using eye
# Create a 2D Identity matrix (1s on the diagonal, 0s elsewhere) of size 5*5
e = torch.eye(5)
print("Using eye ->", e)
```

```python
# Using full
# Creates a 3*3 matrix and fills EVERY position with the number "5".
f = torch.full((3,3),5)
print("Using full ->", f)
```

#### Tensor Shapes..

```python
# Using tensor
# Converts a standard Python list of lists into a PyTorch Tensor.
x = torch.tensor([[7,2,3],[4,5,6]])
x
```

```python
# Using shape
# Checking the shape (dimension) of the tensor.
x.shape
```

```python
# Using empty_like()
# Creates a new tensor with the EXACT same shape and datatype as "x"
# But just like normal empty(), it leaves the memory uninitialized (fills with garbage values).
torch.empty_like(x)
```

```python
# Using zeros_like()
# Creates a new tensor with the EXACT SAME shape and datatype as 'x'.
# Step 1: Allocates the exact same memory space as 'x'.
# Step 2: Manually fills that entire space with 0s (0.0).
torch.zeros_like(x)
```

```python
# Using ones_like()
# Creates a new tensor with the EXACT same shape and datatype as "x".
# Step 1: Allocates the exact same memory space as "x".
# Step 2: Manually fills that entire space with 1s.
torch.ones_like(x)
```

```python
# Using rand_like()
# Step 1: Copies the SHAPE of "x" (2*2).
# Step 2: Fills the space with random decimal numbers between 0.0 and 0.1.
# Step 3: Forces the new tensor to be float 32, IGNORING x's original integer datatype.
torch.rand_like(x, dtype = torch.float32)
```

3. Reduction Operation:-

```python
# Tensor
# Step 1: Create a 2x3 matrix.
# Step 2: Fill it with random integers from 0 to 9.
# Step 3: Forcefully convert those integers into floats (add .0 at the end).
e = torch.randint(size = (2,3), low = 0, high = 10, dtype = torch.float32)
e
```

```python
# Add numbers top-to-bottom (Column-wise sum)
# It crushes 2 rows into 1 single row
torch.sum(e, dim=0)
```

- **dim=0:** This crushes the matrix from Top to Bottom (Vertically). This means it adds the numbers column-wise and crushes all the rows together to form just 1 single line (Row).

- **dim=1:** This crushes the matrix from Left to Right (Horizontally). (This adds the numbers row-wise).

```python
# Add ALL numbers together (No dimension specified)
# It crushes the entire matrix into a single scalar value
torch.sum(e)
```

```python
# Calculate the average of ALL numbers in the matrix
# It divides the total sum by the total number of items
torch.mean(e)
```

```python
# Calculate the average vertically (Top to Bottom)
# It crushes the 2 rows into 1 row of averages
torch.mean(e, dim=0)
```

```python
# Find the middle-most value of the entire matrix
# First it flattens and sorts: [0., 1., 3., 5., 8., 9.]
# Then it picks the middle value (if two values in middle then pick up the small value)
torch.median(e)
```

```python
# Find the smallest number in the entire matrix
# It checks every element and picks the lowest one
torch.min(e)
```

```python
# Find the largest number in the entire matrix
# It checks every element and picks the highest one
torch.max(e)
```

```python
# Multiply ALL numbers inside the matrix together
torch.prod(e)
```

```python
# Calculate the Standard Deviation of all numbers
# It tells us how spread out the numbers are from the mean
torch.std(e)
```

```python
# Calculate the Variance of all numbers
# It tells us the squared spread of numbers from the mean
torch.var(e)
```

```python
# Find the position (index) of the largest number
torch.argmax(e)
```

6. Special Functions:-

```python
# Tensor
k = torch.randint(size = (2,3), low = 0, high = 10, dtype = torch.float32)
k
```

```python
# Calculate the Natural Logarithm element-wise
torch.log(k)
```

- **Strictly Positive Numbers:** The number you are calculating the logarithm for must be greater than zero ($> 0$).
- **If the number is 1:** The answer will always be 0.0 (because $\ln(1) = 0$).
- **If the number is exactly $e$ (2.718):** The answer will be exactly 1.0.
- **If the number is 0:** PyTorch will get upset and return -inf (Negative Infinity).
- **If the number is negative (e.g., -5):** PyTorch will return nan (Not a Number), because the logarithm of negative numbers does not exist.

```python
# Calculate the Exponential element-wise (e^x)
# It will undo the log and bring back our original numbers!
torch.exp(k)
```

- **If the number is 0.0:** The answer will always be 1.0 (because $e^0 = 1$).
- **If the number is 1.0:** The answer will be exactly 2.718 ($e$) (because $e^1 = e$).
- **Undoing Log:** If you first apply log to a number, and then apply exp to the result, you will get the exact original number back! ($e^{\ln(x)} = x$).
- **Always a Positive Answer:** Whether the number inside tensor k is negative or positive, the final answer of torch.exp(k) will always be greater than zero (Positive).

```python
# Calculate the Square Root element-wise
# 4.0 becomes 2.0, 9.0 becomes 3.0, and -9.0 becomes 'nan'
torch.sqrt(k)
```

```python
# Apply the Sigmoid function
# It will squash all these numbers strictly between 0.0 and 1.0
torch.sigmoid(k)
```

- **Large Positive Number:** If the number is much greater than zero (e.g., 10.0 or 100.0), the answer will be squashed very close to 1.0 (e.g., 0.999).

- **Large Negative Number:** If the number is highly negative (e.g., -10.0 or -500.0), the answer will be pushed very close to 0.0 (e.g., 0.0001).

- **Exactly Zero (0.0):** If the number is exactly 0.0, the Sigmoid function will set it to the exact center, which is 0.5.

```python
# Apply Softmax along dimension 0
# It converts the scores into probabilities that sum up to 1.0
torch.softmax(k, dim=0)
```

```python
# Use .clone() to create a completely independent copy
b = a.clone()
```

- **If the id() of two variables is exactly the same:** It means they are not two separate pieces of data, but rather two different names pointing to the exact same data (Aliasing).

- **If you use .clone() or a standard math function (like add()):** The computer creates a completely new memory address, resulting in a completely new and different id().
