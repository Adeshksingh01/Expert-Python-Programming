# Generators

```python
# Create a generator's function
def travel_gen():
  yield "Delhi" # step 1
  yield "Mumbai" # step 2
  yield 2026 # step 3

# Create a generator
my_trip = travel_gen()
```

```python
# Print generator using next()
print(next(my_trip))
```

```python
# Create a generators's function
def Web_startup():
  yield "System Booting" # step 1
  yield "Network Connected" # step 2
  yield "Database Connected" # step 3
  yield "Loading Preferences" # step 4
  yield "Online and Ready" # step 5

# create a generator
web_loading = Web_startup()

# print generator using for loop
for step in web_loading:
  print(step)
```

```python
#Generator expressions provide a compact way to create generators, similar to list comprehensions.
house_price = (area * 5000 for area in range(1000, 1600, 100))

print("1000 sq ft Price is:",next(house_price))
print("1100 sq ft Price is:",next(house_price))
print("1200 sq ft Price is:",next(house_price))
print("1300 sq ft Price is:",next(house_price))
print("1400 sq ft Price is:",next(house_price))
print("1500 sq ft Price is:",next(house_price))

# the expected price of 100 sq ft is 5000
# calculating the price is area * 5000
```

```python
#Generator for factorial series
def factorial_series(n):
  result = 1
  for i in range(1, n+1):
    result *= i
    yield result

# Create a generator
fact = factorial_series(10)

# create a list and print all the factorials in it
print(list(fact))
```

```python
# Create a generator function
def running_average():
    total = 0
    count = 0
    average = 0

    # when condition is true while loop is running
    while True:
        new_value = (yield average)

        if new_value is not None:
            total += new_value
            count += 1
            average = total / count

# Create a generator
avg_cal = running_average()

print("Initial state:", next(avg_cal))

print("After sending 10:", avg_cal.send(10))
print("After sending 20:", avg_cal.send(20))
print("After sending 30:", avg_cal.send(30))
print("After sending None:", avg_cal.send(None))
```
