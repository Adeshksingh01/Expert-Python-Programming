**Creator an Iterator from a list and print it's first two elements using next()?**

```python
# Create a iterator using iter() and next()?
adesh = [1,2,3,4,5] # Iterable list[]
iterator = iter(adesh) # Creating an iterator
```

```python
print(next(iterator))
print(next(iterator))
```

**Create a string into an iterator and print each character using next()?**

```python
text = "Hello"
text_iter = iter(text)
```

```python
print(next(text_iter))
print(next(text_iter))
```

**Create an iterator and print by using for loop?**



```python
data = [1,2,3,4,5,6]
iterator = iter(data)

for item in iterator:
  print(item)
```

**How to you use a for loop to iterate over an iterator**

```python
letter = iter(["A", "B", "C", "D"])

for char in letter:
  print(char)
```

**How do you use iter() and next() to iterate through a range of numbers?**

```python
range_iter = iter(range(10))
```

```python
print(next(range_iter))
print(next(range_iter))
```

```python
print(range(10))
```

**Manually iterate through a tuple using an iterator and a while loop**

```python
fruits = ("Apple", "Banana", "Cherry")
fruit_iter = iter(fruits)

while True:
  try:
    print(next(fruit_iter))
  except StopIteration:
    break
```
