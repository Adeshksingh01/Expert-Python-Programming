Question 1. Create an ElectricCar class that inherits from the Car class and includes an additional attribute, battery_size.


```python
class Car:
  def __init__(self, model, brand):
    self.model = model
    self.brand = brand

class type(Car):
  def __init__(self, model, brand, battery_size):
    super().__init__(model, brand)
    self.battery_size = battery_size
```

```python
car1 = type("M8","BMW","80KW")
print("Car_Model:",car1.model)
print("Car_brand:",car1.brand)
print("Car_battery_size:",car1.battery_size)
```

Question 2. Define a Laptop class that inherits from the Computer class and includes an additional attribute, battery_life.

```python
class Computer:
  def __init__(self, model, brand):
    self.model = model
    self.brand = brand

class laptop(Computer):
  def __init__(self, model, brand, battery_life):
    super().__init__(model, brand)
    self.battery_life = battery_life
```

Question 3. Create a Smartphone class and add another child class of your choice that inherits from an ElectronicDevice class. Include an additional attribute, operating_system.

```python
class Smartphone:
  def __init__(self, model, brand, storage):
    self.model = model
    self.brand = brand
    self.storage = storage

class E_device(Smartphone):
  def __init__(self, model, brand, storage, operating_system):
    super().__init__(model, brand, storage)
    self.operating_system = operating_system
```

```python
Device = E_device("17 pro","Iphone","1 Tb","OIS")
print("Smartphone_model:",Device.model)
print("Smartphone_brand:",Device.brand)
print("Smartphone_storage:",Device.storage)
print("Smartphone_Operating_System:",Device.operating_system)
```

Question 4. Implement a Musician class and create another child class of your choice that extends an Artist class. Include an additional attribute, instrument_played.


```python
class Musician:
    def __init__(self, name, city):
        self.name = name
        self.city = city

class Artist(Musician):
    def __init__(self, name, city, instrument):
        super().__init__(name, city)
        self.instrument = instrument
```

```python
M_info = Artist("AKS", "Lucknow", "Guitar")
print("Musician_name:",M_info.name)
print("Musician_City:",M_info.city)
print("Musician_instrument:",M_info.instrument)
```

Question 5. Define a FictionBook class and add another child class of your choice that inherits from the Book class. Include an additional attribute, genre.

```python
class FictionBook:
  def __init__(self, book, writer):
    self.book = book
    self.writer = writer

class F_Book(FictionBook):
  def __init__(self, book, writer, genre):
    super().__init__(book, writer)
    self.genre = genre
```

```python
Book = F_Book("The Black","AKS","Horror")
print("Book_name:",Book.book)
print("Book_writer:",Book.writer)
print("Book_genre:",Book.genre)
```

Bonus Ques-
Create a student class that takes the name & marks of 3 subjects as arguments in the constructor. Then create a method to print the average.


```python
class Student:
  def __init__(self, name, mark1, mark2, mark3):
    self.name = name
    self.mark1 = mark1
    self.mark2 = mark2
    self.mark3 = mark3

  def avg(self):
    avg_mark = (self.mark1 + self.mark2 + self.mark3) / 3
    return avg_mark
```

```python
Student_1 = Student("AKS", 50, 60, 70)
print(f'Name of Student: {Student_1.name} and Avg marks is: {Student_1.avg()}')
```
